# Wikipedia Scraper

```python
url = input("Wikipedia link: ")
```

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
page = requests.get(url)
```

```python
soup = BeautifulSoup(page.text, 'html')
```

```python
title = soup.title.text.replace("- Wikipedia", "").strip()
```

```python
first_para = soup.find_all('p')[1]
```

```python
print(f"Title : {title} \n")
print(f"First para: {first_para.text.strip()} \n")
```

```python
for link in soup.find_all('a',class_ = 'external text'):
    print(link.get('href'))
print("\n")
```

```python
image = soup.find_all('img')
print("Image links: Captions \n")
for link in image:
    print(f"{link.get('src')}: {link.get('alt')} \n")
    
```
